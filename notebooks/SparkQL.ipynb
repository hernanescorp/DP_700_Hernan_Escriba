## **Creating database objects with Spark**

You can create views and tables. 

```
 df.createOrReplaceTempView("products_view")
```

Create an empty set table:

```
spark.catalog.createTable 
```
Or save an DataFrame object as a table:

```
 df.write.format("delta").saveAsTable("products")
```


**_Fist Step:_** Load and Create a Table

````python
# Upload csv file
df = spark.read.csv("Files/products.csv",
    header = True,
    inferSchema=True)

#Create a Table
df.write.format("delta").saveAsTable("products")
````

## Spark SQL API to data query:

````python
import pyspark.sql

bikes_df = spark.sql("""
SELECT ProductID, ProductName, ListPrice
FROM products
WHERE Category IN ('Mountain Bikes', 'Road Bikes')
""")

display(bikes_df)
````

## **SQL code**

**_First cell line code should be:_** _%%sql_ (magic)

````python
%%sql
SELECT Category, COUNT(ProductID) AS ProductCount
FROM products
GROUP BY Category
ORDER BY Category
````
